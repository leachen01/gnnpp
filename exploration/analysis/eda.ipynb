### Error distribution of sum_stats

```python
print(sum_dfs_train['train'][0].shape)
print(sum_dfs_train['train'][1].shape)
all_t2m_errors = np.array(sum_dfs_train['train'][0]['t2m_mean'] - sum_dfs_train['train'][1]['t2m'])

plt.figure(figsize=(10, 6))
plt.hist(all_t2m_errors.flatten(), bins=100, edgecolor='black')
plt.title('Histogram of t2m errors')
plt.xlabel('t2m value errors (t2m_mean - actual_t2m)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
first_six_days = sum_dfs_train['train'][0][sum_dfs_train['train'][0]['station_id']==0]['time'][:6].to_numpy()
first_six_days
```

```python
# t2m_errors_by day?
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()
first_six_days = sum_dfs_train['train'][0][sum_dfs_train['train'][0]['station_id']==0]['time'][:6].to_numpy()

for i, day in enumerate(first_six_days):
    t2m_errors_day = np.array(sum_dfs_train['train'][0][sum_dfs_train['train'][0]['time'] == day]['t2m_mean'] - sum_dfs_train['train'][1][sum_dfs_train['train'][1]['time'] == day]['t2m'])
    axes[i].hist(t2m_errors_day.flatten(), bins=30, color='skyblue', edgecolor='black')
    axes[i].set_title(pd.to_datetime(day).strftime("%Y-%m-%d"))
    axes[i].set_xlabel("Value")
    axes[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()
e_stations = np.arange(0, 6, 1)

for i, id in enumerate(e_stations):
    t2m_errors_station = np.array(sum_dfs_train['train'][0][sum_dfs_train['train'][0]['station_id']== id]['t2m_mean'] - sum_dfs_train['train'][1][sum_dfs_train['train'][1]['station_id'] == id]['t2m'])
    axes[i].hist(t2m_errors_station.flatten(), bins=30, color='skyblue', edgecolor='black')
    axes[i].set_title(f"Station {id}")
    axes[i].set_xlabel("Value")
    axes[i].set_ylabel("Frequency")

for ax in axes:
    ax.set_xlim(-10, 10)
    ax.set_ylim(0, 600)
plt.tight_layout()
plt.show()
```

### Altitude vs. orography

```python

from sklearn.metrics import mean_squared_error, mean_absolute_error
print(f"alt mean: {t_stations_f['altitude'].mean()}, orog mean: {t_stations_f['orog'].mean()}")
print(f"alt std: {t_stations_f['altitude'].std()}, orog std: {t_stations_f['orog'].std()}")
print(f"alt max: {t_stations_f['altitude'].max()}, orog max: {t_stations_f['orog'].max()}")
print(f"alt min: {t_stations_f['altitude'].min()}, orog min: {t_stations_f['orog'].min()}")
rmse = np.sqrt(mean_squared_error(t_stations_f['altitude'], t_stations_f['orog']))
mae = mean_absolute_error(t_stations_f['altitude'], t_stations_f['orog'])
print(f"RMSE: {rmse}, MAE: {mae}")
t_stations_f
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(t_stations_f.index, t_stations_f['altitude'], color='blue', label='Actual Altitude')
plt.scatter(t_stations_f.index, t_stations_f['orog'], color='red', label='Model Orography')
plt.xlabel('Station')
plt.ylabel('Elevation (m)')
plt.title('Comparison of Actual Altitude and Model Orography')
plt.legend()
plt.show()
```
